* yes24 파이썬 검색
* 도서명 / 가격 / 점수 
* 점수 9.5 이상인 것 20권만 화면 출력

* 브라우저 실행

```python
driver = webdriver.Chrome()
driver.maximize_window()
```

```python
url = 'http://www.yes24.com/Main/default.aspx'
headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.69 Safari/537.36'}
driver.get(url)
```

* 검색창 단어 입력 및 검색

```python
elem = driver.find_element_by_class_name("iptTxt")
elem.send_keys("파이썬")
```

```python
elem.send_keys(Keys.ENTER)
```

* 도서명 크롤링

* 사이트 정보 불러오기

```python
soup = BeautifulSoup(driver.page_source)
```

```python
t= soup.find_all('a', attrs = {"class":"gd_name"})
for i in range(len(t)):
    title = t[i].get_text()
    print(title)
```

* 별점 체크

```python
t_2 = soup.find_all('div', attrs = {'class':'info_row info_rating'})
for i in range(len(t_2)):
    score = t_2[i].get_text()
    print(score)
```
